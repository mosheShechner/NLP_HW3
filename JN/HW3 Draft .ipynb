```python
from nltk.corpus import LazyCorpusLoader, BracketParseCorpusReader
from nltk import Tree
from nltk.tree import ParentedTree


treebank = LazyCorpusLoader('treebank/combined', BracketParseCorpusReader, r'wsj_.*\.mrg')

i = 3

print("\nbefore")
print(treebank.parsed_sents()[i])

sent = treebank.parsed_sents()[i].copy()

def simplify_functional_tag(tag):
    if '-' in tag:
        tag = tag.split('-')[0]
    return tag

def scanTree(node):
    for n, child in enumerate(node):
        if isinstance(child, Tree):
            #print("node %s" %child.label())
            #print("node simplified %s" % simplify_functional_tag(child.label()))
            if (child.label() == "-NONE-"): 
                print("found NONE")
                child.set_label("DEL")
            else:
                child.set_label(simplify_functional_tag(child.label()))
                scanTree(child)
        else:
            continue 
            #print(child)
            
scanTree(sent)


print("\nafter")
print(treebank.parsed_sents()[i])
```

```python
def deleteNoneNodes(sent):
    for sub in sent.subtrees():
        for n, child in enumerate(sub):
            if isinstance(child, str):
                continue
            if len(list(child.subtrees(filter=lambda x:x.label()=='-NONE-')))==len(child.leaves()):
                del sub[n]
```
